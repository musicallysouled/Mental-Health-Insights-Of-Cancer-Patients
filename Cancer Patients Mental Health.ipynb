# Insight to Mental Health of Cancer Patients

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mental Health Dataset.csv')
```

```python
df.head()
```

```python
df.isnull().any()
```

```python
df['intensity'].value_counts()
```

```python
df.groupby(['predicted','intensity']).count()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='intensity', data=df, palette = 'mako')
plt.title('Intensity of Posts')
plt.xlabel('Intensity')
plt.ylabel('Count')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
check_titles = ''.join(map(str,df['posts']))
wordcloud_title = WordCloud(width = 800, height = 800,
                            background_color ='white',
                            min_font_size = 10).generate(check_titles)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(wordcloud_title)
plt.axis('off')
plt.title('Cancer Patients post sentiment', fontsize=20)
plt.tight_layout(pad=2)
plt.show()
```
